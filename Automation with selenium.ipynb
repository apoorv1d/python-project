```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time

website = 'https://www.adamchoi.co.uk/goalsbyhalf/detailed'
path = '/usr/bin/chromedriver'

# Create a Service object with the path to ChromeDriver
service = Service(executable_path=path)

driver = webdriver.Chrome(service=service)
driver.get(website)

# Locating the button with the help of XPath
all_matches_button = driver.find_element(By.XPATH, '//label[@analytics-event="All matches"]')
# Click the button
all_matches_button.click()

#selecting the country from the dropdown to make the dataframe dynamic

dropdown = Select(driver.find_element(By.ID,'country'))
dropdown.select_by_visible_text('Spain')

#This pauses the execution of the code for 3 sec before the execution of further code
time.sleep(3)


# Extracting data from the table
matches = driver.find_elements(By.TAG_NAME, 'tr')

date = []
home_team = []
score = []
away_team = []

# find_elements returns a list unlike find_element that returns a single value
for match in matches:
    cells = match.find_elements(By.TAG_NAME, 'td')
    if len(cells) > 0:
        date.append(cells[0].text)
        home_team.append(cells[1].text)
        score.append(cells[2].text)
        away_team.append(cells[3].text)

# Printing extracted data
for i in range(len(date)):
    print(f"{date[i]} - {home_team[i]} vs {away_team[i]}: {score[i]}")

driver.quit()

#Converting the data into a dataframe
df = pd.DataFrame({'date':date, 'home_team': home_team, 'score':score, 'away_team':away_team})
df.to_csv('football_data.csv',index = False)
print(df)
```
